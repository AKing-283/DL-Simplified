```python
import torch
from IPython.display import Image, display

# Define the path to the weights file and image file
weights_path = '/home/arismita/ML/yolov5_training/yolov5/landCover/yolov5s_size640_epochs5000_batch16_small/weights/best.pt'
img_path = '/home/arismita/ML/RetinaNet/landcover/dataset/JPEGImages/aachen_10.jpg'

# Load the model
model = torch.hub.load('ultralytics/yolov5', 'custom', path=weights_path)

# Perform inference on an image
results = model(img_path)

# Display the results
results.print()

# Specify the directory where you want to save the results image
save_dir = '/home/arismita/ML/yolov5_training/yolov5/landCover/runs/detect/exp4'
results.save(save_dir)  # save results to specified directory

# Display the saved image
image_path = f'{save_dir}/aachen_10.jpg'  # Construct the path to the saved image
display(Image(filename=image_path))
```
